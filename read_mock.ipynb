```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gname="/Users/Canev001/Documents/GWlensing/GitHub/mock1.txt" 
mock_z,mock_mu = np.genfromtxt(gname,dtype="float", \
comments="#", usecols=(0,1), unpack=True)


plt.plot(mock_z,mock_mu,'.')
plt.ylabel(r'$\mu$', fontsize=20)
plt.xlabel('z', fontsize=20)
plt.show()


plt.hist(mock_z, bins=20, density='True', histtype='step')
plt.ylabel('dn/dz', fontsize=20)
plt.xlabel('z', fontsize=20)
plt.show()
```
